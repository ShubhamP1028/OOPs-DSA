<h3><center><b>Class And Object</b></center></h3>

<b>Class :</b>

A class is a blueprint or a template for creating objects. It defines a set of attributes (variables) and methods (functions) that the objects created from it will possess

<b>Object</b>

An object is an instance of a class. It's a concrete entity created based on the class's blueprint, with its own unique set of attribute values

```python
# Defining a classic
class Dog:
    # Class attribute
    species = "Dog"

    # Constructor method (initializes object attributes)
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

    # Instance method
    def bark(self):
        return f"{self.name} says Woof!"
```

```python
# Creating objects of the Dog class
my_dog = Dog("Tommy", 3)
another_dog = Dog("Lucy", 5)

# Accessing object attributes
print(f"{my_dog.name} is a {my_dog.species} and is {my_dog.age} years old.")
print(f"{another_dog.name} is a {another_dog.species} and is {another_dog.age} years old.")

# Calling object methods
print(my_dog.bark())
print(another_dog.bark())
```
